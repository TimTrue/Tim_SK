6. Напишите запрос, позволяющий вычислить, какую зарплату для DS в **среднем** указывают для каждого типа требуемого опыта (уникальное значение из поля *experience*). 

При решении задачи примите во внимание следующее:
1. Рассматриваем только вакансии, у которых заполнено хотя бы одно из двух полей с зарплатой.
2. Если заполнены оба поля с зарплатой, то считаем зарплату по каждой вакансии как сумму двух полей, делённую на 2. Если заполнено только одно из полей, то его и считаем зарплатой по вакансии.
3. Если в расчётах участвует null, в результате он тоже даст null (посмотрите, что возвращает запрос select 1 + null). Чтобы избежать этой ситуацию, мы воспользуемся функцией [coalesce](https://postgrespro.ru/docs/postgresql/9.5/functions-conditional#functions-coalesce-nvl-ifnull), которая заменит null на значение, которое мы передадим. Например, посмотрите, что возвращает запрос `select 1 + coalesce(null, 0)`

```python
# текст запроса
query_6_6 = f'''
    SELECT
        distinct experience as "Тип опыта",
        /* Считаем среднюю ЗП используя coalesce() для замены null на 0 */
        round(AVG(coalesce((salary_from+salary_to)/2, salary_from, salary_to)),0) as "Средняя зарплата"
    FROM public.vacancies
    where (salary_from is not null or salary_to is not null) and
        (name ilike '%data scientist%' or
        name ilike '%data science%' or
        name ilike '%исследователь данных%' or
        name ilike '%machine learning%' or
        name ilike '%машинн%обучен%' or
        (name not like '%HTML%' and name like '%ML%'))
        GROUP BY 1
'''
```

7.2. Выяснить в какой сфере деятельности задействовано наибольшее количество вакансий

```python
# текст запроса
query_7_2 = f'''
    SELECT
        i.name as "Сфера деятельности деятельности",
        count(distinct v.employer_id) as "Количество вакансий"
    FROM public.industries i
    LEFT JOIN public.employers_industries ei ON i.id = ei.industry_id
    LEFT JOIN public.vacancies v ON ei.employer_id = v.employer_id
    GROUP BY 1
    ORDER BY 2 DESC
'''
```

```python
# результат запроса
df = pd.read_sql_query(query_7_2, connection)
df
```

7.3.1. Выяснить среднюю зарплату по региону Москва и сравнии его с средним значением ЗП по всем вакансиям.

```python
# текст запроса
query_7_3_1 = f'''
    (SELECT
        (round(avg(v.salary_from)) + round(avg(v.salary_to)))/2 as "Среднее значение ЗП",
        'Москва' as "Выборка"
    FROM public.vacancies v
    LEFT JOIN public.areas a ON a.id = v.area_id
    WHERE a.name = 'Москва')
    UNION ALL
    (SELECT
        (round(avg(v.salary_from)) + round(avg(v.salary_to)))/2 as "Среднее значение ЗП",
        'Все вакансии' as "Выборка"
    FROM public.vacancies v)
'''
```

```python
# результат запроса
df = pd.read_sql_query(query_7_3_1, connection)
df
```

Вывод

Средняя зарплата выше в Москве (137001.5) чем в среднем по всем регионам (90801.0) на 46200 или на 50%.

7.3.2. Определить рейтинг работодателй по количеству вакансий в регионе.

```python
# текст запроса
query_7_3_2 = f'''
    SELECT
        e.name as "Работодатель",
        count(v.id) as "Количество вакансий"
    FROM public.employers e
    LEFT JOIN public.vacancies v ON e.id = v.employer_id
    LEFT JOIN public.areas a ON a.id = v.area_id
    WHERE a.name = 'Москва'
    GROUP BY e.name
    ORDER BY 2 DESC
'''
```

```python
# результат запроса
df = pd.read_sql_query(query_7_3_2, connection)
df
```

На первом местре по количеству вакансий в регионе Москва находится Сбер. Рейтинг работодателей по количеству вакансий в Москве значительно отличается от рынка в целом.

7.3.3. Выяснить тип рабочего графика, наиболее популярного в регионе.

```python
# текст запроса
query_7_3_3 = f'''
    SELECT
        v.schedule as "Тип рабочего графика",
        v.employment as "Тип трудоустройства",
        count(v.id) as "Количество вакансий"
    FROM public.vacancies v
    LEFT JOIN public.areas a ON a.id = v.area_id
    WHERE a.name = 'Москва'
    GROUP BY schedule, employment
    ORDER BY 3 DESC
'''
```

```python
# результат запроса
df = pd.read_sql_query(query_7_3_3, connection)
df
```

Вывод    

Требования к типу трудоустройства и типу рабочего графика в соотношении с количеством вакансий не отличаются от рынка в целом.
