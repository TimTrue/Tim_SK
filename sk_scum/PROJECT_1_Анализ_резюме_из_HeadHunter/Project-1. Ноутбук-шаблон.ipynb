1. Начнем с простого - с признака **"Образование и ВУЗ"**. Его текущий формат это: **<Уровень образования год выпуска ВУЗ специальность...>**. Например:
* Высшее образование 2016 Московский авиационный институт (национальный исследовательский университет)...
* Неоконченное высшее образование 2000  Балтийская государственная академия рыбопромыслового флота…
Нас будет интересовать только уровень образования.

Создайте с помощью функции-преобразования новый признак **"Образование"**, который должен иметь 4 категории: "высшее", "неоконченное высшее", "среднее специальное" и "среднее".

Выполните преобразование, ответьте на контрольные вопросы и удалите признак "Образование и ВУЗ".

Совет: обратите внимание на структуру текста в столбце **"Образование и ВУЗ"**. Гарантируется, что текущий уровень образования соискателя всегда находится в первых 2ух слов и начинается с заглавной буквы. Воспользуйтесь этим.

*Совет: проверяйте полученные категории, например, с помощью метода unique()*


```python
def get_education_level(education):
    # Создаем список категорий образования
    level_list = ['высшее', 'неоконченное высшее', 'среднее специальное', 'среднее']
    # Разбиваем строку на слова через пробелы
    education_list = education.lower().split(' ')
    # Проверяем соответсвтие категории образования сначала в первых двух словах
    if str(education_list[0] + ' ' + education_list[1]) in level_list:
        education_type =  str(education_list[0] + ' ' + education_list[1])
    # Затем только в первом
    else:
        education_type = education_list[0]
    return education_type

# Применяем функцию
hh_database['Образование'] = hh_database['Образование и ВУЗ'].apply(get_education_level)

# Проверяем количество уникальных категорий
print('Количество уникальных категорий образования: ', hh_database['Образование'].nunique())

#Удаляем столбцец "Образование и ВУЗ"
hh_database.drop(['Образование и ВУЗ'], axis=1, inplace=True)

# Проверяем состав столбцов
print('Список столбцов: ', hh_database.columns)
```

3. Следующим этапом преобразуем признак **"Опыт работы"**. Его текущий формат - это: **<Опыт работы: n лет m месяцев, периоды работы в различных компаниях…>**. 

Из столбца нам необходимо выделить общий опыт работы соискателя в месяцах, новый признак назовем "Опыт работы (месяц)"

Для начала обсудим условия решения задачи:
* Во-первых, в данном признаке есть пропуски. Условимся, что если мы встречаем пропуск, оставляем его как есть (функция-преобразование возвращает NaN)
* Во-вторых, в данном признаке есть скрытые пропуски. Для некоторых соискателей в столбце стоит значения "Не указано". Их тоже обозначим как NaN (функция-преобразование возвращает NaN)
* В-третьих, нас не интересует информация, которая описывается после указания опыта работы (периоды работы в различных компаниях)
* В-четвертых, у нас есть проблема: опыт работы может быть представлен только в годах или только месяцах. Например, можно встретить следующие варианты:
    * Опыт работы 3 года 2 месяца…

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
def get_experience(arg):
# Список обозначений месяцев
    months_exclude_list = ['месяц', 'месяца', 'месяцев']
# Список обозначений лет
    years_exclude_list = ['год', 'лет', 'года']
# Разбиваем строку на слова через пробел    
    experience_list = str(arg).split(' ')
# Проверяем потенциальное указание на месяц и год
    if arg is np.nan or experience_list[0] != 'Опыт':
        return np.nan
    else:
        output = 0
        if (experience_list[5] in months_exclude_list) and (experience_list[3] in years_exclude_list):
    # Складываем месяцы неполного и полных лет
            output = int(experience_list[4]) + (int(experience_list[2])*12)
        elif (experience_list[5] in months_exclude_list):
    # Учитываем только месяцы неполного года
            output = int(experience_list[4])
    # Учитываем только месяцы полных лет
        elif (experience_list[3] in years_exclude_list):
            output = int(experience_list[2])*12
        return output

# Применяем функцию
hh_database['Опыт работы (месяц)'] = hh_database['Опыт работы'].apply(get_experience)

# Удаляем столбцец "Образование и ВУЗ"
hh_database.drop(['Опыт работы'], axis=1, inplace=True)
```

    Еще один важный факт: при выгрузки данных у некоторых соискателей "потерялась" информация о готовности к командировкам. Давайте по умолчанию будем считать, что такие соискатели не готовы к командировкам.
    
Выполните преобразования и удалите столбец **"Город, переезд, командировки"** из таблицы.

*Совет: обратите внимание на то, что структура текста может меняться в зависимости от указания ближайшего метро. Учите это, если будете использовать порядок слов в своей программе.*


```python
def get_move_info(arg):
    # Создаем множество городов-миллиоников
    million_cities = {'Новосибирск', 'Екатеринбург', 'Нижний Новгород', 'Казань', 'Челябинск', 'Омск', 'Самара', 'Ростов-на-Дону', 
                      'Уфа', 'Красноярск', 'Пермь', 'Воронеж', 'Волгоград' }
    # Создаем множество "главных" городов, являющие и значениями требуемых категорий
    main_cities = {'Москва', 'Санкт-Петербург'} 
    # Создаем множество негативны ключевых слов (в данном случае только частица "не")
    negative_list = {'не'}
    # Создаем множество ключевых слов для проверки наличия данных о готовности к командировкам
    trip_list = {'командировкам', 'командировке', 'командировки', 'командировкам'}
    # Разбиваем строку на слова через заяпятую
    info_list = arg.split(' , ')
    
    # Определяем город проживания
    if set(info_list[0].split()) & main_cities:
        city = info_list[0]
    elif set(info_list[0].split()) & million_cities:
        city = 'город-миллионник'
    else:
        city = 'другие'
    
    # Определяем готовность к переезду и коандировкам
    if set(info_list[-1].split()) & trip_list:
        if set(info_list[-1].split()) & negative_list:
            trip_ready = False
        else:
            trip_ready = True
        if set(info_list[-2].split()) & negative_list:
                move_ready = False
        else:
                move_ready = True
    else:
        trip_ready = False
        if set(info_list[-1].split()) & negative_list:
                move_ready = False
        else:
                move_ready = True
       
    return pd.Series([city, move_ready, trip_ready])

# Применяем функцию    
hh_database[['Город', 'Готовность к переезду', 'Готовность к командировкам']] = hh_database['Город, переезд, командировки'].apply(get_move_info)

# Удаляем столбцец "Город, переезд, командировки"
hh_database.drop(['Город, переезд, командировки'], axis=1, inplace=True)  
```

5. Рассмотрим поближе признаки **"Занятость"** и **"График"**. Сейчас признаки представляют собой набор категорий желаемой занятости (полная занятость, частичная занятость, проектная работа, волонтерство, стажировка) и желаемого графика работы (полный день, сменный график, гибкий график, удаленная работа, вахтовый метод).
На сайте hh.ru соискатель может указывать различные комбинации данных категорий, например:
* полная занятость, частичная занятость
* частичная занятость, проектная работа, волонтерство
* полный день, удаленная работа
* вахтовый метод, гибкий график, удаленная работа, полная занятость

Такой вариант признаков имеет множество различных комбинаций, а значит множество уникальных значений, что мешает анализу. Нужно это исправить!

Давайте создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

Такой метод преобразования категориальных признаков называется One Hot Encoding и его схема представлена на рисунке ниже:
<img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/ohe.jpg>
Выполните данное преобразование для признаков "Занятость" и "График", ответьте на контрольные вопросы, после чего удалите их из таблицы

```python
def get_employment(employment):
    # Создаем список ключевых слов для проверки категории занятости
    employment_categories = ['полная занятость', 'частичная занятость', 'проектная работа', 'стажировка', 'волонтерство']
    # Разбиваем строку на слова через заяпятую
    employment_set = employment.split(', ')
    employment_set_var = set(employment_set)
    
    # Определяем категории занятости
    if employment_set_var & set(employment_categories[0].split(',')):
        full_time = True
    else:
        full_time = False
    
    if employment_set_var & set(employment_categories[1].split(',')):
        part_time = True
    else:
        part_time = False
        
    if employment_set_var & set(employment_categories[2].split(',')):
        project = True
    else:
        project = False
        
    if employment_set_var & set(employment_categories[3].split(',')):
        internship = True
    else:
        internship = False
        
    if employment_set_var & set(employment_categories[4].split(',')):
        volunteering = True
    else:
        volunteering = False
       
    return pd.Series([full_time, part_time, project, internship, volunteering])


def get_schedule(schedule):
    # Создаем список ключевых слов для проверки категории рабочего графика
    schedule_categories = ['вахтовый метод', 'гибкий график', 'удаленная работа', 'полный день', 'сменный график']
    # Разбиваем строку на слова через заяпятую
    schedule_set = schedule.split(', ')
    schedule_set_var = set(schedule_set)
        
    # Определяем категории рабочего графика 
    if schedule_set_var & set(schedule_categories[0].split(',')):
        watch = True
    else:
        watch = False
    
    if schedule_set_var & set(schedule_categories[1].split(',')):
        flexible = True
    else:
        flexible = False
        
    if schedule_set_var & set(schedule_categories[2].split(',')):
        remotely = True
    else:
        remotely = False
        
    if schedule_set_var & set(schedule_categories[3].split(',')):
        full_day = True
    else:
        full_day = False
        
    if schedule_set_var & set(schedule_categories[4].split(',')):
        shifting = True
    else:
        shifting = False    
       
    return pd.Series([watch, flexible, remotely, full_day, shifting])
    
hh_database[['полная занятость', 'частичная занятость', 'проектная работа', 'стажировка', 'волонтерство']] = hh_database['Занятость'].apply(get_employment)

hh_database[['вахтовый метод','гибкий график', 'удаленная работа', 'полный день', 'сменный график']] = hh_database['График'].apply(get_schedule)

# Удаляем столбцец "Город, переезд, командировки"
hh_database.drop(columns=['Занятость', 'График'], axis=1, inplace=True) 
```

```python
h = hh_database.iloc[49:151]
h.head(2)

def get_payment(data):
    currency_dict = \
    {'грн.':'UAH', 'руб.':'RUB', 'KZT':'KZT', 'USD':'USD', 
    'бел.руб.':'BYN', 'EUR':'EUR', 'сум':'UZS', 
    'KGS':'KGS', 'AZN':'AZN'}
    salary_split = data.split()
    for i in currency_dict.keys():
        if i == salary_split[1]:
            return pd.Series([currency_dict.get(i), salary_split[0]])

h[['валюта','зарплата']]= h['ЗП'].apply(get_payment)

h.head(2)

merged = h.merge(
    ExchangeRates,
    left_on = ['Обновление резюме','валюта'],
    right_on = ['date','currency'],
    how='left'
)
merged[['currency', 'proportion', 'close']].fillna(1)
merged['currency'].apply(lambda x: 1 if x == np.nan else x)

display(merged)



# def get_transfer(data):
#     for 

# for row in h.items():
#     salary_split = h.loc['ЗП'].split()
#     print(salary_split)
#     for i in currency_dict.keys():
#         if i == salary_split[1]:
#             print(currency_dict.get(i))
#             h['валюта'] = currency_dict.get(i)
#     print(h)
#     merged = hh_database.merge(
#         ExchangeRates,
#         left_on='валюта',
#         right_on= 'currency',
#         how='left'
#     )
#     display(merged)        
```
